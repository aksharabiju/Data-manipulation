```python
data['Height'] = np.random.uniform(150, 180, size=len(data))
data
```

```python
x=data['Number'].mean()
data['Number'].fillna(x,inplace=True) 
data
```

```python
x=data['Weight'].mean()
data['Weight'].fillna(x,inplace=True) 
data
```

```python
x=data['Salary'].mean()
data['Salary'].fillna(x,inplace=True) 
data
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data.dropna(subset=['College'],inplace=True)
data
```

```python
data.dropna(subset=['Salary'],inplace=True)
data
```

```python
position_groups = data.groupby('Position')['Name'].apply(list)

for position, names in position_groups.items():
    print(f"Employees in {position}:")
    for name in names:
        print(name)
    print()
```

```python
age_bins = [20, 30, 40, 50, 60, 70, 100]  
age_labels = ['20-30', '31-40', '41-50', '51-60', '61-70', '70+'] 

data['AgeGroup'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)
age_group_counts = data['AgeGroup'].value_counts()
most_common_age_group = age_group_counts.idxmax()

print(f"Age group most employees belong to: {most_common_age_group}.")
```

```python
team_position_salary = data.groupby(['Team', 'Position'])['Salary'].sum()

max_spending_team_position = team_position_salary.idxmax()
max_spending_salary = team_position_salary.max()

print("Team and Position with the highest spending:")
print(f"Team: {max_spending_team_position[0]}")
print(f"Position: {max_spending_team_position[1]}")
print(f"Total Salary: {max_spending_salary}")
```

```python
import matplotlib.pyplot as plt

correlation = data['Age'].corr(data['Salary'])
print(f'Correlation between Age and Salary:{correlation}')

plt.figure(figsize=(10, 12))
plt.scatter(data['Age'], data['Salary'], alpha=1.0)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
